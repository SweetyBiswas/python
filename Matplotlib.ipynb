# ===========Pie chart===========

# line graph useful when u have continous variable

# but for categorical variable......

# suppose we have male and female in my data...and want to see the proportion(Ratio....anupat,a part,share or number considered in comparative relation to a whole )of males and females in my data

# To show proportion====we use Pie chart

### pie() function use kori

```python
lab="python",'c++','c','java'
a=np.array([215,130,245,210])
plt.pie(a,labels=lab)

#labels parameter used to create lable
```

### Change Colour

```python
lab="python",'c++','c','java'
a=np.array([215,130,245,210])
col=['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(a,labels=lab,colors=col)
```

### explode parameter

```python
lab="python",'c++','c','java'
a=np.array([215,130,245,210])
col=['gold','yellowgreen','lightcoral','lightskyblue']
ex=(0.1,0,0,0)
plt.pie(a,labels=lab,colors=col,explode=ex)
```

```python
lab="python",'c++','c','java'
a=np.array([215,130,245,210])
col=['gold','yellowgreen','lightcoral','lightskyblue']
ex=(0.1,0,0,0.3)
plt.pie(a,labels=lab,colors=col,explode=ex)
```

### shadow parameter


#### Bydefault false tyke....true korle shadow dakhabe

```python
lab="python",'c++','c','java'
a=np.array([215,130,245,210])
col=['gold','yellowgreen','lightcoral','lightskyblue']
ex=(0.1,0,0,0)
plt.pie(a,labels=lab,colors=col,explode=ex,shadow=True)
```

### =========== autopct parameter ===============

####  1.1 means point r por akta digit show korbe,
#### 1.2 mane point r por 2 to digit show korbe

```python
lab="python",'c++','c','java'
a=np.array([215,130,245,210])
col=['gold','yellowgreen','lightcoral','lightskyblue']
ex=(0.1,0,0,0)
plt.pie(a,labels=lab,colors=col,explode=ex,shadow=True,autopct='%1.1f%%')
```

```python
lab="python",'c++','c','java'
a=np.array([215,130,245,210])
col=['gold','yellowgreen','lightcoral','cyan']
ex=(0.1,0,0,0)
plt.pie(a,labels=lab,colors=col,explode=ex,shadow=True,autopct='%1.2f%%')
```
